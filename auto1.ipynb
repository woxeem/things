```python
options = webdriver.ChromeOptions()
options.headless = True

# S = lambda X: driver.execute_script('return document.body.parentNode.scroll' + X)
driver = webdriver.Chrome(options=options)
driver.implicitly_wait(6)
# driver.set_window_size(S('Width'), S('Height'))
driver.set_window_size(1000, 2100)

temp_dir = tempfile.TemporaryDirectory()

wb_source = load_workbook('/test1.xlsx')
wb_dest = load_workbook('/test1.xlsx')

sheets = [sheet for sheet in wb_source.sheetnames if re.search('\d+', sheet)]

for i, ticket in enumerate(tqdm(sheets), start = 1):
    
    ws_source = wb_source[ticket]

    try:

        ticketID = int(ws_source['A21'].value)

    except Exception:

        continue

    if (i == 1):

        USER = input('Enter your login')

        PASSWORD = input('Enter your password')
        
        driver.get(f'https://{USER}:{PASSWORD}@{BASE_URL}{ticketID}')

        code = input('Enter verification code')

        driver.find_element(By.ID, INPUT_ID).send_keys(code)

        driver.find_element(By.CSS_SELECTOR, SIGN_TAG).click()

    else:

        driver.get(f'https://{BASE_URL}{ticketID}')

    time.sleep(1.75)

    # print(f'Progress: {i}/{len(sheets)}')

    img_path = f'{temp_dir.name}/screenshot_{ticketID}.png'

    driver.find_element(By.TAG_NAME, 'body').screenshot(img_path)

    img = Image(img_path)

    img.anchor = 'F3'

    ws_dest = wb_dest[ticket]
    ws_dest.add_image(img)

wb_dest.save('/final.xlsx')

temp_dir.cleanup()
```
